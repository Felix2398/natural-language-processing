## Task 5

Draw the syntax trees for the following sentences using the Penn Treebank Syntactic tagest introduced in the [lecture](https://aclanthology.org/J93-2004.pdf) (see table 3).

(a) He buys a house. 

```
S ( 
    NP ( 
        PRP(He)
        VP ( 
            VBZ(Buys)
            NP( 
                DT(a)
                NN(house)
            )
        )    
    )
)
```

(b) Marburg is a hilly city. 

```
S ( 
    NP ( 
        NNP(Marburg)
    )
    VP ( 
        VBZ(is)
        NP ( 
            DT(a)
            JJ(hill)
            NN(city)
        )
    )
)
```
(c) He loves her but she doesn’t love him back.

```
S (
    S (
        NP (
            PRP(he)
        )
        VP (
            VBZ(loves)
            NP (
                PRP(her)
            )
        )
    )
    CC(but)
    S (
        NP (
            PRP(she)
        )
        VP (
            VBZ(does)
            RB(not)
            VP (
                VB(love)
                NP (
                    PRP(him)
                )
                ADVP (
                    RB(back)
                )
            )
        )
    )
)
```
