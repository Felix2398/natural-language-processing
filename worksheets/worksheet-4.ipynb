## Task 3

Implement a Na√Øve Bayes Classifier in Python from scratch.

(a) It should be able to be trained on a CSV file in the following format:

```
sentence,class
"The quick brown fox jumps over the lazy dog.", A
"Lorem ipsum dolor sit amet, consectetur adipiscing elit.", B

```

Use the Pandas library to read the CSV. Implement add-1 smoothing and add the unknown word to the vocabulary.

````python
import pandas as pd

def clean(sentence):
    return sentence.replace('.', '').replace(',', '').lower().split()

def train(csv_file):
    df = pd.read_csv(csv_file)
    classes = df['class'].unique().tolist()
    priors_map = {key: 0 for key in classes}
    word_map = {key: [] for key in classes}
    count_map = {key: {} for key in classes}
    sum_map = {key: 0 for key in classes}
    vocabulary = set()

    for row in df.itertuples(index=False):
        words = clean(row[0])
        vocabulary.update(words)
        word_map[row[1]].extend(words)
        priors_map[row[1]] += 1

    for clazz in classes:
        for word in vocabulary:
            count = word_map[clazz].count(word)
            count_map[clazz][word] = count + 1
            sum_map[clazz] += count
        sum_map[clazz] += len(vocabulary)

    return {'classes': classes, 'rows': len(df), 'vocabulary': vocabulary, 'class_counts': priors_map, 'word_count': count_map, 'overall_count': sum_map}
````

(b) Write a function predict that takes a string as input and predicts the a class based on a classifier trained in (a).

````python
def predict(model, sentence):
    words = clean(sentence)
    predictions = {}

    for clazz in model['classes']:
        probability = model['class_counts'][clazz] / model['rows']

        for word in words:
            count = 1
            if word in model['vocabulary']:
                count = model['word_count'][clazz][word]
            probability *= count / (model['overall_count'][clazz])
        
        predictions[clazz] = probability
    
    return predictions
````

````python
from io import StringIO

csv_content = """sentence,class
"fun, couple, love, love", comedy
"fast, furious, shoot", action
"couple, fly, fast, fun, fun", comedy
"furious, shoot, shoot, fun", action
"fly, fast, shoot, love", action
"""

csv_file = StringIO(csv_content)

model = train(csv_file)
print('Model: ', model)
print('Prediction: ', predict(model, 'fast, couple, shoot, fly'))
````
